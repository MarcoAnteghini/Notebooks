## PTS1 Search 

```python
import re
```

```python
#listed below are two possible PTS1 consensus motif, past the in the re.search() function to change
#[SACHEQ]-[KRH]-[LAF] yeast
#[ASCNPHTG]-[RKHQNSL]-[LMIVF] other eucaria (e.g. zebrafish)

fasta_sequences='zeb_prot_c-term.fasta'
#fasta file containing the sequences of interest, in this case the C-terminal part of the zebrasifh proteome

output='zeb_matches.txt'
#arbitrary name of the output file

with open(fasta_sequences) as fasta, open(output,'w') as result:
    for line in fasta:
        if line[0] == '>':
            idx=line.rstrip()[1:]   
        else:
            match = re.search(r'[ASCNPHTG][RKHQNSL][LMIVF]$', line)
            if match:
                result.write(idx+' '+line)  
```

```python
#print of the matching sequences. In the first column we can see the uniprot ID, 
#in the second column the last 3 aminoacids corresponding to the PTS1 signal
with open('zeb_matches.txt') as m:
    for line in m:
        print(line.rstrip())
```
